## 모험가들의 레벨업 도전

각 캐릭터는 힘, 민첩, 지능, 체력이라는 능력치를 가지고 있습니다. 이 능력치들을 이용해 다양한 도전을 해결하고 사냥을 통해 경험치를 얻어 레벨업을 해야 합니다.

도적: 힘(40), 민첩(80), 지능(30), 체력(60)

__[문제 1: 첫 번째 사냥]__

마법사는 몬스터를 사냥하기 위해 마법을 사용합니다.

마법을 사용하려면 지능이 50 이상이어야 합니다.

마법사의 지능이 50 이상인지 확인하세요.

```python
# 마법사: 힘(10), 민첩(20), 지능(80), 체력(50)
Wizard_power = 10
Wizard_speed = 20
Wizard_brain= 80
Wizard_hp = 50

Wizard_brain >= 50
```

<br>

__[문제 2: 두 번째 사냥]__

전사는 강력한 공격을 하기 위해 자신의 힘을 사용합니다.

전사의 힘이 60 이상이고 체력이 80 이상인지 확인하세요.

```python
# 전사: 힘(80), 민첩(30), 지능(20), 체력(100)
Worrior_power = 80
Worrior_speed = 30
Worrior_brain= 20
Worrior_hp = 100

Worrior_power >= 60 and Worrior_hp >= 80
```

<br>

__[문제 3: 세 번째 사냥]__

도적은 민첩함을 이용해 몬스터를 회피하고 공격합니다.

도적의 민첩이 70 이상이거나 힘이 50 이상인지 확인하세요.

```python
# 도적: 힘(40), 민첩(80), 지능(30), 체력(60)
Rogue_power = 40
Rogue_speed = 80
Rogue_brain = 30
Rogue_hp = 60

Rogue_speed >= 70 or Rogue_power >= 50
```

<br>

__[문제 4: 네 번째 사냥]__

힐러는 파티의 체력을 회복시키기 위해 마법을 사용합니다.

힐러의 지능이 60 이상이고, 체력이 50 이상인지 확인하세요.

```python
# 힐러: 힘(20), 민첩(40), 지능(70), 체력(70)

Healer_power = 20
Healer_speed = 40
Healer_brain = 70
Healer_hp = 70

Healer_brain >= 60 and Healer_hp >= 50
```

<br>

__[문제 5: 레벨업]__

모든 캐릭터가 사냥을 성공적으로 마쳤습니다. 이제 경험치를 얻어 레벨업을 해야 합니다.


각 캐릭터의 경험치가 100 이상인지 print()를 이용해 확인해주세요.

<br>

경험치는 아래와 같습니다.

마법사: 120

전사: 150

도적: 110

힐러: 90



```python
Wizzard_exp = 120
Worrior_exp = 150
Rogue_exp = 110
Healer_exp = 90

print("Wizzard의 경험치는 100이상입니까?",Wizzard_exp >= 100)
print("Worrior의 경험치는 100이상입니까?",Worrior_exp >= 100)
print("Rogue의 경험치는 100이상입니까?",Rogue_exp >= 100)
print("Healer의 경험치는 100이상입니까?",Healer_exp >= 100)
```
