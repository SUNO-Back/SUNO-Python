__[문제7]__

변수명 전화번호에 들어 있는 번호를 모두 붙여서 출력되도록 코드를 작성해주세요.

<br>

phone = "010-1234-5678"

출력 예시 : "01012345678"

<br>

__Tips__
- split() 함수는 문자열을 특정 구분자를 기준으로 나누어 리스트로 반환합니다.


```python
phone = "010-1234-5678"
a = phone.split("-")
result = "".join(a)
print(result)
```
