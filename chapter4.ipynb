##사칙연산
더하기 : +

빼기 : -

곱하기 : *

나누기 :
*  / : (예 5/2 = 2.5) 정확한 나누기
*  // : (예5/2 = 2) 몫을 출력
*  % : (예 5/2=1) 나머지를 출력

거듭제곱 : **

```python
# 더하기
print(3+2)
```

```python
# 빼기
print(3-1)
```

```python
# 곱하기
print(3*2)
```

```python
# 나누기 /
print(7/2)

# //
print(7//2)

# %
print(7%2)
```

내장함수 : divmod

> 내장함수 : 자주 사용하는 기능을 미리 만들어 놓은것을 내장함수라 함.

**divmod** 목과 나머지를 동시에 구해줌

```python
print (divmod(3,2))
print (divmod(7,7))
print (divmod(13,2))
print (divmod(19,4))
print (divmod(15,5))

print(7//4,7%4)
```
