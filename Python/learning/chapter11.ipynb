##시퀀스 자료형
*문자열:  
    * 시퀀스형에서 사용할 수 있는 메소드 (count(), index(), split() 등,,..)  
    * 슬라이싱 ([:],[::])  
    * 요소들로 이루어져 가능했던 기능들 (요소 합치기, 요소찾기 등등)  
*list :  
    * 시퀀스형에서 사용할 수 있는 메소드 (count(), index(), split() 등,,..)  
    * 슬라이싱 ([:],[::])  
    * 요소들로 이루어져 가능했던 기능들 (요소 합치기, 요소찾기 등등)  
    * list 지원하는 다양한 메소드(append(), insert(), del, sort(), reverse())  
*tuple :   
    * 리스트처럼 요소를 일렬로 저장하지만 안에 저장된 요소를 변경, 추가, 삭제할 수 없다.  
    * 값을 추가, 삭제, 변경할 수 없기 때문에, 사용하지 못하는 메소드가 굉장히 많다.  
    * tuple의 생성 방법은 tuple = 값, 또는 tuple = (값,값) 두가지가 있다
    * 그리고 엄청난 비밀이 숨겨져 있다
    * 튜플은 소괄호 입니다.


```python
# 리스트 = [], 튜플 = ()
튜플 = (1,"0|",3.0)
print(type(튜플))
print(튜플)

# 리스트 = [1,"0|", 3.0]
# print(type(리스트))
# print(리스트)

튜플2 = 1, "0|", 3
print(type(튜플2))
print(튜플2)

# 하나의 값만 들어있는 튜플만 만들고 싶다면?
튜플_솔로 = (1,)
print(type(튜플_솔로))
print(튜플_솔로)

튜플_나는솔로 = (1)
print(type(튜플_나는솔로))
print(튜플_나는솔로)
```

```python
# 값 추가, 삭제 , 변경불가 합니다.
print(튜플)
print(dir(튜플))

튜플_리스트화 = list(튜플)
print(type(튜플_리스트화))
```

```python
# 정말로 값의 삽입, 삭제, 변경을 지원하지 않는지

#튜플.append("두리안")  추가 안됨.
#del 튜플[2]          삭제 안됨.
튜플[1] = "너만 믿는다" 변경안됨
print(튜플)
```

```python
# 번외
수강생 = "중꺽그마" # 중요한건 꺾여도 그냥 하는 마음..

수강생_list = list(수강생)
수강생_tuple = tuple(수강생)

print(수강생_list)
print(수강생_tuple)

수강생_list[1] = "뿌"
print(수강생_list)

#join함수 공부 알고 있으면 코테(코딩테스트)에 좋다
취뽀_수강생 = "".join(수강생_list)
print(취뽀_수강생)
```

```python
# 튜플과 리스트의 공통점
print(튜플)
#길이를 구할 수 있다
len(튜플)
#곱하기 연산으로 요소를 복사할 수 있다
print(튜플 * 3)
#튜플은 요소들끼리 덧셈 할 수 있다
print(튜플 + 튜플)
#튜플은 인덱싱이 가능하다
print(튜플[2])
#튜플의 인덱스
튜플.index(3.0)
#튜플 요소의 개수
튜플.count(1)
#튜플도 슬라이싱이 가능하다
print(튜플[1:3])
```
