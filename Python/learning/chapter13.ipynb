* set

        * set은 수학의 집합을 의미합니다.
        * 딕셔너리와 같이 순서가 없습니다.
        * 값의 중복을 허락하지 않습니다. (아주 중요함)
        * set 생성 방법도 중괄호 {}
        * dict는{키:밸류}, set은{값} 만 입력

```python
채소 = {"당근","배추","대파","양파"}
type(채소)
print(채소)

# 순서가 없고, 출력하면 요소가 다르게 나오는것도 특징입니다.

# 세트를 만드는 다양한 방법
세트 = set()
세트
세트2 = {}
세트2
```

```python
# 중복을 제일 싫어하는 set

사과_리스트 = list("apple")
print(사과_리스트)
사과_튜플 = tuple("apple")
print(사과_튜플)
사과_set = set("apple")
print(사과_set)
```

```python
# set = 집합 , 합집합,교집합,차집합 기타 등등~~!
# 대표적으로 합집합, 교집합, 차집합

채소1 = {"당근","양파","오이","배추"}
채소2 = {"양파","오이","대파","가지"}

#합집합(union)
채소 | 채소2
set.union(채소1,채소2)
#교집합(intersection)
채소 & 채소2
set.intersection(채소1,채소2)


#차집합(difference)
채소1 - 채소2
set.difference(채소1,채소2)
```
