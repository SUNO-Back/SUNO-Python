### 변수에 값을 넣는것 => 할당
## 변수 : 하나의 값을 저장할 수 있는 공간을 생성하는 것

sum = 1 + 2 => 1+2의 계산된 결과를 sum이라는 변수에 할당한다.  
sum은 변수명이고 "=" 같다는 의미가 아니라 할당(assignment)이라고 합니다.  
"=" : 할당 연산자  
"+" : 산술 연산자  
+= : 산술 후 바로 할당하는 문법  
/=, -=,*= 모두 가능합니다  

```python
#변수를 사용하면서 사용 가능한 문법

sum = 1+2
sum = sum + 3
sum = sum + 4
sum = sum + 5

print(sum)

sum2 = 1 + 2
sum2 +=3
sum2 +=4 = sum = sum + 4
sum2 +=5

print(sum2)
```

```python
#파머님이 내주시는 문제
#붕어빵 철이 돌아왔죠
#붕어팡(팥) : 2000원, 붕여빵(슈크림) 2500원, 붕어팡 (잡채) : 3000원
#무인 주문 기계를 만들어주세요
#붕어빵(잡채)3개,붕어팡(슈크림)2개를 주문하면
#"주문이 완료되었습니다. 붕어빵(슈크림)2개, 붕어빵(잡채)3개 총 5개 결제 금액은 14000원 입니다"
# 이 출력되도록 코드를 작성해주세요
#변수를 만들때 스네이크 케이스에 언더바는 2개 / print()문의 포메팅은 f-string

#변수값 입력
bread__p, bread__s, bread__j = "붕어빵(팥)","붕어빵(슈크림)","붕어빵(잡채)"
bread__p__sell,bread__s__sell,bread__j__sell = 2000,2500,3000
order_s,order_j = 2,3

#붕어빵 계산값
order_s__pay,order_j__pay = bread__s__sell * 2,bread__j__sell *3
#붕어빵 구매 개수
total = order_s + order_j
#붕어빵 총 가격
total_pay = order_s__pay + order_j__pay
#출력
print(f'주문이 완료되었습니다{bread__s}{order_s}개,{bread__j}{order_j}개 총{total}개 결제 금액은 {total_pay}원 입니다')
```
