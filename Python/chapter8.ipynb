```python
# 본격적인 논리연사자
# True : 1 / False : 0 / and : *
True and True # 1 * 1 = 1
True and False # 1 * 0 = 0
False and True # 0 * 1 = 0
False and False # 0 * 0 = 0
```

```python
# 본격적인 논리연사자
# True : 1 / False : 0 / or : +
True or True # 1 + 1 = 1      수학이 아니다 True + True = True다
True or False # 1 + 0 = 1
False or True # 0 + 1 = 1
False or False # 0 + 0 = 0
```

```python
# not : 반전

not True # = Fasle

not False # = True
```

```python
# 점점 복잡해지는 조건들 !!!

10 == 12 or 6 != 2 # True
2 > 5 and 7 == 7 # False
not 14 <= 15  # False
not 7 is 7.0 # True
```

```python
# 굉장히 중요한 내용
# 단락 평가 - 조건적으로 앞에까지만 평가해서 출력값을 내는것. ex) False and, True or
print(False and True)
print(False and "힘들어요")
print(False or "힘들어요")
# False and 는 뒤에 뭐가 나오든 무조건 False가 나온다

print(True and False)
print(True and "힘들어요") # True and 는 뒤에 뭐가 나오는지 보고 그 값을 출력
print(True or "힘들어요") # True or 은 뒤에 뭐가 나오든 무조건 True로 나온다
```
