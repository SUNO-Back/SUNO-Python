* dict(딕셔너리)

        *list,tuple,str,range(범위안에 연속된 데이터를 만드는 시퀀스)
        -얘네들에게는 공통적인 특징이 있다. 데이터의 연관성이 없다.
        *dict는 여러개의 값이 일렬로 정렬되면서도 값끼리의 연관성이 존재
        (예시 : 게임 캐릭터 힘:30, 지능:20, 체력:40, 민첩90)
        -dict 생성 방법은 중괄호 형이다.

```python
# list = [] tuple = () dict = {}

원펀맨 = {"힘":30, "지능":20, "체력":40, "민첩":90}
print(원펀맨)

# (키 : 값)
# (key1:value1, key2:value2)

# 시퀀스 데이터는 인덱싱을 이용해 데이터를 찾았습니다.
원펀맨["힘"]

print(f'원펀맨님은 힘이{원펀맨["힘"]}입니다.')
```

```python
# 딕셔너리를 만드는 다양한 방법

#빈 딕셔너리 만들기
딕셔너리 = []
딕셔너리

딕셔너리2 = dict()
딕셔너리2

#값을 넣어 만드는 방법
#dict()로 생성시 값 넣는 기호 "",: 주의!
딕셔너리3 = dict(힘=30, 지능=20, 체력=40, 민첩=90)
print(딕셔너리3)

#zip() : 병렬처리를 도와줌. ( 이건 따로 더 공부 해보기 )
딕셔너리4 = dict(zip(["힘","지능","체력","민첩"],[30,20,40,90]))
딕셔너리4
```

```python
# 값의 중복은 허락되는데, 키의 중복은 허락하지 않는다.

원펀맨 = {"힘":30, "지능":20, "체력":40, "민첩":90,"힘":40,"지구력":90} # 여기서 "힘" = key 30 = values 가장 뒤에 입력된게 최신이라 생각하고 뒤에껄 입력
원펀맨

"체력" in 원펀맨
```

```python
주문_개수 = map(int, input(f'붕어빵 맛 순서대로 개수를 입력해주세요 안살꺼면 0을 입력해주세요(팥,슈크림,잡채)').split(','))

메뉴별_가격 = {"팥":2000,"슈크림":2500,"잡채":3000}

주문_개수 = list(주문_개수)
총_주문_개수 = sum(주문_개수)
total = 0

total += 메뉴별_가격["팥"] * 주문_개수[0]
total += 메뉴별_가격["슈크림"] * 주문_개수[1]
total += 메뉴별_가격["잡채"] * 주문_개수[2]

print(f'주문이 완료되었습니다. 팥{주문_개수[0]}개, 슈크림{주문_개수[1]}개, 잡채{주문_개수[2]}개,  총{총_주문_개수}개 결제 금액은 {total}원 입니다')


```
