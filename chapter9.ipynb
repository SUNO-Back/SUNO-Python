```python
넌센스, 정답 = input().split(",") # split()으로 입력하면 공백으로 나눠서 입력 '에스파를 만나면' < 에스파를 = 넌센스 , 만나면 = 정답 으로 인식됨 (",") 은 , 기준으로 나눈다는 뜻
print(f'넌센스 퀴즈 : {넌센스}')
print(f'정답 : {정답}')
```

```python
#시퀀스 자료형에 다 지원되는 메소드
번호1, 번호2, 번호3, 번호4 = input().split(",")

번호1 = int(번호1)
번호2 = int(번호2)
번호3 = int(번호3)
번호4 = int(번호4)

print(type(번호1), 번호2, 번호3, 번호4)
```

```python
#map()
번호1, 번호2, 번호3, 번호4 = map(int, input().split(",")) # map안에 들어오는 값들을 int(정수)로 변환 하고싶으면 map()안에 바꾸고싶은 변환타입을 맨 앞으로 기입해야함.

print(type(번호1), 번호2, 번호3, 번호4)
```

```python
# 레거시 : 누군가 남겨두고간 코드
# 리팩토링 : 결과의 변경 없이 코드의 구조를 재조정 하는 것.
# input()

#bread__p, bread__s, bread__j = "붕어빵(팥)","붕어빵(슈크림)","붕어빵(잡채)"
bread__p, bread__s, bread__j = input('붕어빵(팥),붕어빵(슈크림),붕어빵(잡채)를 순서대로 개수를 입력하고 꼭 쉼표를 입력하고 안사는 붕어빵에는 0을 넣어줘').split(",")
bread__p__sell,bread__s__sell,bread__j__sell = 2000,2500,3000
order_s,order_j = 2,3
order_s__pay,order_j__pay = bread__s__sell * 2,bread__j__sell *3
total = order_s + order_j
total_pay = order_s__pay + order_j__pay
print(f'주문이 완료되었습니다{bread__s}{order_s}개,{bread__j}{order_j}개 총{total}개 결제 금액은 {total_pay}원 입니다')
```
