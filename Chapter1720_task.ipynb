__[문제1]__

print_even 리스트 내에 있는 값 중 짝수만 화면에 출력되도록 print_score 함수를 생성 후 결과값을 출력해주세요

print_even ([1, 3, 2, 10, 12, 11, 15])

```python
print_even = ([1, 3, 2, 10, 12, 11, 15])

class print_score:
    def __init__(self,nums):
        self.nums = nums

    def numbers(self):
        for i in self.nums:
            if i % 2 == 0:
                print(i)


total = print_score(print_even)
total.numbers()

```

__[문제6]__

5번 문제에서 생성한 인스턴스의 이름, 나이, 성별을 출력해주세요.



출력 결과  ) <br>
이름 : 김코  
나이 : 30

__Tips__


- 인스턴스 변수에 접근하여 값을 출력하면 됩니다  : )

```python
class oz:
    def __init__(self,name,age,score):
        self.name = name
        self.age = age
        self.score = score



coding = oz("김코",30,"A")

print("이름:",coding.name)
print("나이:",coding.age)
print("코딩레벨:",coding.score)
```

__[문제7]__

오즈 (oz) 클래스에서 이름, 나이, 성별을 출력하는 who() 메소드를 추가하고 who()메소드를 이용해 아래 출력 결과를 만들어주세요


출력 결과 ) <br>
이름: 김코, 나이: 30, 코딩레벨: A

```python
class oz:
    def __init__(self,name,age,level):
        self.name = name
        self.age = age
        self.level = level


    def who(self):
        print(f"이름: {self.name}, 나이:{self.age} 코딩레벨:{self.level}")

coding = oz("김코",30,"A")

coding.who()
```
